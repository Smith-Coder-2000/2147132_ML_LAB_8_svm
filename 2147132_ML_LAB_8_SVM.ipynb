```python
df = pd.read_csv("/content/drive/MyDrive/kag_risk_factors_cervical_cancer.csv")
#prints first 10 file in csv
df.head(40)
```

```python
for feature in df.columns:
    df[feature].replace('?',np.nan,inplace=True )
    df[feature].fillna(value=0,inplace=True)
for feature in df.columns:
    df[feature].replace(0,df[feature].median(),inplace=True)
df.head()
```

```python
df.replace(to_replace=np.nan,value=0000)
```

```python
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import LabelEncoder
Le = LabelEncoder()
```

```python
features = ['Age', 'Dx:CIN', 'Dx:HPV','Dx:Cancer']
xx=features
X = df[features]
y = df['Biopsy']
print(X)
print(y)
```

```python
#spliting data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=0)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

from sklearn.metrics import accuracy_score
```

```python
# SVC with linear kernel
C = 1.0 # SVM regularization parameter
# LinearSVC (linear kernel)
svc =SVC(kernel='linear', C=C)
svc.fit(X_train, y_train)
y_pred1 = svc.predict(X_test)
print ('Accuracy-svm', accuracy_score(y_test, y_pred1))
```

```python
# SVC with RBF kernel
rbf_svc =SVC(kernel='rbf', gamma=0.8, C=C)
rbf_svc.fit(X_train, y_train)
y_pred2 = rbf_svc.predict(X_test)
print ('Accuracy-svm', accuracy_score(y_test, y_pred2))
```

```python
# SVC with polynomial (degree 3) kernel
poly_svc =SVC(kernel='poly', degree=3, C=C)
poly_svc.fit(X, y)
y_pred3 = poly_svc.predict(X_test)
print ('Accuracy-svm:', accuracy_score(y_test, y_pred3))
```
